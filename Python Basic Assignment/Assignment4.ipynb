17. 

copy.copy() returns shallow copy. In this case original elements do get change when new list is updated because reference of list is copied to new list.

copy.deepcopy() returns deep copy. In this case original elements are not changed when there is a modification on new list. This happens because copy of list is given to new list.


But this concept is only applicable for compound objects like list, tuple.

```python
import copy
```

```python
#deep copy
l3 = [1,2,[9,1],3]
#before update
print('l3 before',l3)
#deep copy
l4 = copy.deepcopy(l3)

#update
l4[2][1] = 9

#updated list l4
print('l4',l4)

#orginal list l3
print('l3 after',l3)
```

```python
#shallow copy
l3 = [1,2,[9,1],3]
#before update
print('l3 before',l3)
#shallow copy
l4 = copy.copy(l3)

#update
l4[2][1] = 9

#updated list l4
print('l4',l4)

#orginal list l3
print('l3 after',l3)
```
