5 How many global scopes are there in a Python program? How many local scopes?


Local scope(functional): In this type of scope variable has a life of the function call which means that as soon as the function is called local variable is created and when the function call is complete the varaible is destroyed automatically. Its scope is limited to that function block only and it will throw an error when called outside the function.

Enclosing scope(non-local scope) : In this case there is a nesting of functions i.e one function inside another function. Here the inner function and it's variables are only visible to outer function and the inner function comes to forefront just when outer function is called. The enclosing variables are present inside the outer function but outside the inner function. We can use nonlocal keyword to modify it. 

module scope (global scope) : In this case variable has global scope. We can use global keyword to modify its value. The scope of this type of variable is not confined to a function and it can be used through out the program.

Buil_in : Example of these are keywords like print(), for, break, continue...

The LEGB rule is followed for precedence.

Local then enclosed then global and at last built_in.

```python
#global variable
y=10
def func1():
  global y
  y = 11
  #local variable
  x = 10
  print('x value is',x)
  print('y value is',y)
print('y value is',y)

func1()
```

```python
#enclosed

def func2():
  y =10
  print('y value before modification',y)
  def func3():
    #enclosed variable
    x =11
    nonlocal y
    y = y+1
    print('x value is',x)
    print('y value after modification',y)
  
  func3()

func2()
```
