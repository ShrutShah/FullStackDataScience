```python
#1.Write a Python Program to Find the Factorial of a Number?

def factorial(n):
  l=1
  for i in range(1,n+1):l*=i
  return l

factorial(5)
```

```python
factorial(1)
```

```python
#2 Write a Python Program to Display the multiplication Table?

def multiply(n):
  for i in range(1,n+1):
    print ('{} * {} = {}'.format(n,i,n*i))

multiply(10)
```

```python
multiply(15)
```

```python
#3. Write a Python Program to Print the Fibonacci sequence?

def fibonacci(n):
  a =1
  b=1
  l=[]
  l.append(b)
  for i in range(1,n):
    a, b = b, a+b
    l.append(a)
  print(l)  
```

```python
d = int(input('Enter number of terms'))
fibonacci(d)
```

```python
#4.Write a Python Program to Check Armstrong Number?

def armstrong(n):
  l=[]
  n1=n
  while n%10 != 0:
    r= n%10
    l.append(r)
    n =int(n/10)
  c=0
  for i in l:
    c+=i**3
  
  if c == n1:
    print('Armstrong')

  else :
    print('Not Armstrong') 

armstrong(123)
```

```python
armstrong(153)
```

```python
#5.Write a Python Program to Find Armstrong Number in an Interval?

def armstrong(n):
  l=[]
  n1=n
  while n != 0:
    r= n%10
    l.append(r)
    n =int(n/10)
  c=0
  for i in l:
    c+=i**3
  
  if c == n1:
    return 1

  else :
    return 0
r=[]
for j in range(500):
  if armstrong(j):
    r.append(j)
```

```python
#6.Write a Python Program to Find the Sum of Natural Numbers?
def natural(n):
  sum=0
  for i in range(1,n+1):
    sum += i
  return sum
```

```python
# sum of first 10  natural no.s
natural(10) 
```

```python
# sum of first 25  natural no.s
natural(25)
```
