```python
#1.Write a Python program to convert kilometers to miles?
def distance():
  d=float(input('Enter Distance in kms :'))
  m = 0.621371*d
  return m
m1 = distance()
print('distance in miles :',m1)
```

```python
#2.Write a Python program to convert Celsius to Fahrenheit?

def temp():
  c = float(input('Enter temp in Celsius'))
  f = (1.8*c) + 32
  print('Temp in Fahrenheit :',f) 

temp()
```

```python
# 4. Write a Python program to solve quadratic equation?

def quad():
  a=float(input('Enter square coefficient : '))
  b=float(input('Enter linear coefficiuent : '))
  c=float(input('Enter constant : '))

  root1 = (-b + math.sqrt(b*b - 4*a*c))/(2*a)
  root2 = (-b - math.sqrt(b*b - 4*a*c))/(2*a)

  return root1,root2

root1, root2 = quad()
print('The roots of quadratic equations are {} and {}'.format(root1,root2))

```

```python
#5. Write a Python program to swap two variables without temp variable?

def swap(no1,no2):
  no1, no2 = no2, no1
  return no1, no2

a = float(input("Enter no1 "))
b = float(input('Enter no2 '))
no1, no2 = swap(a,b)
print('After Swap the numbers are {} and {}'.format(no1, no2))
```
