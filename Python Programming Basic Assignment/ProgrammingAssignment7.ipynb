```python
# 1 Write a Python Program to find sum of array?

arr = [1,2,3,4,5]
s=0
for i in arr:
  s+=i
print('Sum =',s)
```

```python
# 2.Write a Python Program to find largest element in an array?

arr=[1,4,9,32,99,23]
max = 0
for i in arr:
  if i > max:
    max=i

print('max = ',max)
```

```python
#5.Write a Python Program to check if given array is Monotonic?

m = [1,2,2,3,4,5,6,6,7]

for i in range(len(m)):
  if i!= len(m)-1:  
    if m[i] <= m[i+1]:
      continue
if i == len(m)-1:
  print('Monotonic')
else:
  print('Not monotonic')
```

```python
m = [1,2,5,3,4,5,6,6,7]

for i in range(len(m)):
  if i!= len(m)-1:  
    if m[i] <= m[i+1]:
      continue
    else:
      break
if i == len(m)-1:
  print('Monotonic')
else:
  print('Not monotonic')
```
